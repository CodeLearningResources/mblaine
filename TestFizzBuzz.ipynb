```python
# Fizz Buzz Program
# If num %3 and num % 5, print fizzbuzz
for num in range(1 , 101):
    print(num)
    if (num % 3 == 0):
        if (num % 5 == 0):
            print("FizzBuzz")
        else:
            print("Fizz")
    elif (num % 5 == 0):
        print("Buzz")
    else:
        print("none")
```
